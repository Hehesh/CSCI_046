Think of the premise of binary search as a search for a specific element within a sorted list in O($\sqrt{n}$) time.

### Creating test cases

Test cases verify edge cases.

```python
cards = [10, 9, 8, 2, 1]
query = 2
output = 3
```

Alternatively, make a dictionary to map inputs to outputs.

```python
test_1 = {
    'input': {
        'cards': cards,
        'query': query
    },
    'output': output
}
```

To use the test case, you can write a logical statement.

```python
locate_card(**test_1['input']) == test_1['output']
```

### Determining edge cases

Even from this simple problem, there may be multiple edge cases:
- The query is in the middle of the list
- The query is at the beginning of the list
- The query is at the end of the list
- The list contains just one element: query
- The list does not contain query
- The list is empty
- The list contains repeating numbers (including or excluding query)

```python
# Query at the beginning
test_2 = {
    'input': {
        'cards': cards,
        'query': cards[0]
    },
    'output': 0
}

# Query at the end
test_3 = {
    'input': {
        'cards': cards,
        'query': cards[-1]
    },
    'output': len(cards) - 1
}

# List contains just query
test_4 = {
    'input': {
        'cards': [query],
        'query': query
    },
    'output': 0
}

# List does not contain query
test_5 = {
    'input': {
        'cards': cards,
        'query': 7
    },
    'output': -1
}

# List is empty
test_6 = {
    'input': {
        'cards': [],
        'query': query
    },
    'output': -1
}

# Repeating numbers
test_7 = {
    'input': {
        'cards': [8, 8, 6, 6, 6],
        'query': 8
    },
    'output': [2, 3, 4]
}

test_cases = [test_1, test_2, test_3, test_4, test_5, test_6, test_7]
```

```python
def locate_card(cards, query):
    index = 0
    while True:
        if index == len(cards):
            return -1

        if cards[index] == query:
            return index
        
        index += 1
```

```python
for test in test_cases:
    print(locate_card(**test['input']) == test['output'], f"expected {test['output']} but got {locate_card(**test['input'])}")
```
